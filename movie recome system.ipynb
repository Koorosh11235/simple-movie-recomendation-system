```python
import pandas as pd
```

```python
crdf = pd.DataFrame(pd.read_csv("./credits.csv"))
modf = pd.DataFrame(pd.read_csv("./movies.csv"))
```

```python
modf.head(1)
```

```python
df = modf.merge(crdf, on = "title")
df.head(2)
```

```python
df.info()
```

```python
df = df[["movie_id", "title", "overview", "genres", "keywords","cast","crew"]]
df.head(1)
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
import ast
def converision(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
df["genres"] = df["genres"].apply(converision)
df
```

```python
def converision_cast(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            l.append(i["name"])
            counter+= 1
        else:
            break
        return l
```

```python
df["cast"] = df["cast"].apply(converision_cast)
```

```python
df.head(5)
```

```python
df["keywords"] = df["keywords"].apply(converision)
df.head(2)
```

```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            l.append(i["name"])
    return l
```

```python
df["director"] = df["crew"].apply(fetch_director)
```

```python
df = df.drop(["crew"], axis = 1)
```

```python
df.head(2)
```

```python
"""df= df.dropna()"""
```

```python
df["overview"] = df["overview"].apply(lambda x : x.split())
df.head(2)
```

```python
def replacing(x):
    if x is None:
        l = []
    else:
        l = [i.replace(" ", "") for i in x]
    return l
df["genres"] = df["genres"].apply(lambda x : [i.replace(" ", "") for i in x])
df["keywords"] = df["keywords"].apply(lambda x : [i.replace(" ", "") for i in x])
df["cast"] = df["cast"].apply(replacing)
df["director"] = df["director"].apply(lambda x : [i.replace(" ", "") for i in x])
```

```python
df["tags"] = df["genres"]+df["cast"]+df["director"]+df["keywords"]
df.head(2)
```

```python
sdf = df[["movie_id", "title", "tags"]]
```

```python
sdf["tags"] = df["tags"].apply(lambda x : ", ".join(x))
sdf
```

```python
sdf["tags"] = sdf["tags"].apply(lambda x : x.lower())
sdf
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000, stop_words="english")
```

```python
vectors = cv.fit_transform((sdf["tags"])).toarray()
```

```python
vectors[0]
```

```python
(cv.get_feature_names_out())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split(", "):
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
sdf["tags"] = sdf["tags"].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key= lambda x : x[1])[1:6]
```
